```python
import pandas as pd
df = pd.read_csv('overlap_comparison_dutch_sub.csv')
print(df.shape)
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
tfidf = []
mprank = []
labels = []

rows = df['idx'].max()
for idx in range(rows):
    for key,row in df[df['idx']==idx].iterrows():
        algo = row[1]
        keywords = row[2:].dropna().tolist()
        if algo=='multipartiterank':
            mprank.append(keywords)
        elif algo == 'tfidf':
            tfidf.append(keywords)
        else:
            labels.append(keywords)
```

```python
tfidf_intersect_mprank = []
for tf,mp in zip(tfidf,mprank):
    tfidf_intersect_mprank.append(set(tf) & set(mp))
```

```python
mp_intersect_labels = []
for label,mp in zip(labels,mprank):
    mp_intersect_labels.append(set(label) & set(mp))
```

```python
tfidf_intersect_labels = []
for label,tf in zip(labels,tfidf):
    tfidf_intersect_labels.append(set(label) & set(tf))
```

```python
overlap_size_tfidf_to_mprank = 0
overlap_size_mprank_to_tfidf = 0

mp_label_min_tfidf = 0
tfidf_label_min_mp = 0


for idx in range(rows):
    try:
        overlap_size_tfidf_to_mprank+= (len(tfidf_intersect_mprank[idx])/len(tfidf[idx]))/rows
    except:
        pass
    
    try:
        overlap_size_mprank_to_tfidf+= (len(tfidf_intersect_mprank[idx])/len(mprank[idx]))/rows
    except:
        pass
    
    try:
        mp_label_min_tfidf+= (len(mp_intersect_labels[idx] - set(tfidf[idx]))/len(mp_intersect_labels[idx]))/rows
    except ZeroDivisionError:
        pass
    
    try:
        tfidf_label_min_mp+= (len(tfidf_intersect_labels[idx] - set(mprank[idx]))/len(tfidf_intersect_labels[idx]))/rows
    except ZeroDivisionError:
        pass
    
    
    
    
    
        
```

```python
print(f'tfidf is for {overlap_size_tfidf_to_mprank*100:.2f}% a subset of mprank:')
print(f'mprank is for {overlap_size_mprank_to_tfidf*100:.2f}% a subset of tfidf:')

print(f'correct answer overlap of mprank with tfidf {(1-mp_label_min_tfidf)*100:.2f}%')
print(f'correct answer overlap of tfidf with mprank {(1-tfidf_label_min_mp)*100:.2f}%')


```
